```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import mysql.connector

#data sources
data_2018 = "Resources/top2018.csv"
data_2017 = "Resources/top2017.csv"
```

```python
data_2017 = pd.read_csv(data_2017)
data_2017.head(5)
```

```python
clean_2017 = data_2017[['name', 'artists','danceability','energy','speechiness','liveness','tempo','duration_ms']]
cleaner_2017 = clean_2017.rename(columns = {"name": "song_name", "duration_ms": "duration_min"})
cleaner_2017['duration_min'] = cleaner_2017['duration_min'].div(60000).round(2)
cleaner_2017['tempo'] = cleaner_2017['tempo'].round(2)
cleaner_2017['danceability'] = cleaner_2017['danceability'].multiply(100)
cleaner_2017['energy'] = cleaner_2017['energy'].multiply(100)
cleaner_2017['speechiness'] = cleaner_2017['speechiness'].multiply(100)
cleaner_2017['liveness'] = cleaner_2017['liveness'].multiply(100)
cleaner_2017
```

```python
data_2018 = pd.read_csv(data_2018)
data_2018.head()
```

```python
clean_2018 = data_2018[['name', 'artists','danceability','energy','speechiness','liveness','tempo','duration_ms']]
cleaner_2018 = clean_2018.rename(columns = {"name": "song_name", "duration_ms": "duration_min"})
cleaner_2018['duration_min'] = cleaner_2018['duration_min'].div(60000).round(2)
cleaner_2018['tempo'] = cleaner_2018['tempo'].round(2)
cleaner_2018['danceability'] = cleaner_2018['danceability'].multiply(100)
cleaner_2018['energy'] = cleaner_2018['energy'].multiply(100)
cleaner_2018['speechiness'] = cleaner_2018['speechiness'].multiply(100)
cleaner_2018['liveness'] = cleaner_2018['liveness'].multiply(100)
cleaner_2018
```

```python
#link jn to pg
connection_string = "postgres:postgres@localhost:5432/etl_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
cleaner_2017.to_sql(name='data_2017', con=engine, if_exists='append', index=False)
```

```python
cleaner_2018.to_sql(name='data_2018', con=engine, if_exists='append', index=False)
```

```python
#Formatting tables
clean_2018 = data_2018[['name', 'artists','danceability','energy','speechiness','liveness','tempo','duration_ms']]
cleaner_2018 = clean_2018.rename(columns = {"name": "song_name", "duration_ms": "duration_min"})
cleaner_2018['duration_min'] = cleaner_2018['duration_min'].div(60000).round(2)
cleaner_2018['tempo'] = cleaner_2018['tempo'].round(2)
cleaner_2018 = cleaner_2018.style.format({
    'danceability': '{:,.1%}'.format, 
    'energy': '{:,.1%}'.format,
    'liveness': '{:,.1%}'.format,
    'speechiness': '{:,.1%}'.format,
})
cleaner_2018
```
